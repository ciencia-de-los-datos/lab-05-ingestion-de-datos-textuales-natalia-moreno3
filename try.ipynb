```python
import pandas as pd
import glob
import fileinput
import os
```

```python
def load_input(input_directory, folder_name):

    input_directory = input_directory + os.sep + folder_name
    secuence = []
    file_names = glob.glob(input_directory + "/*")
    with fileinput.input(files=file_names) as f:
        for line in f:
            secuence.append(
                [line, folder_name]
            )
    
    return secuence
```

```python
def make_data(folder_type):

    folders = ['positive', 'neutral', 'negative']
    data = []
    for folder in folders:
        lines = load_input(f'data{os.sep}{folder_type}', folder)
        data = data + lines

    df = pd.DataFrame(data, columns=['phrase', 'sentiment'])
    df.to_csv(f'{folder_type}_dataset.csv', index=False)

    return df
```
